# Load data from visited places

You can download your data from https://takeout.google.com/settings/takeout?pli=1

```
file_path = '/content/gdrive/My Drive/MincienciatonProject_Data/'
file_name = '2020_JUNE.json'
with open(file_path + file_name) as f:
    data = json.load(f)
```

```
places = []
visits = []

for data_unit in data["timelineObjects"]:
    if "activitySegment" in data_unit.keys():
        visits.append(activitySegment(data_unit["activitySegment"]))
    elif "placeVisit" in data_unit.keys():
        places.append(placeVisit(data_unit["placeVisit"]))
    else:
        print("Error")

df_pl = pd.DataFrame(places)
df_vs = pd.DataFrame(visits)

df_pl.drop([0,3,6], axis = 1, inplace = True)
df_pl.columns = range(df_pl.shape[1])
df_pl.columns = ['lat','lon','ti','tf']
df_pl['time'] = pd.to_datetime(df_pl.iloc[:,3]) - pd.to_datetime(df_pl.iloc[:,2])

df_pl.head()
```

```
m=folium.Map(location=[5.070275, -75.513817])
for i in range(df_pl.shape[0]):
  folium.Marker(location=[df_pl.iloc[i,0], df_pl.iloc[i,1]],popup='Default popup Marker1').add_to(m)
m
```
